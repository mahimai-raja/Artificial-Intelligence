# Gradient Descent 

    Y = mx + c
    
    Gradient Descent is an algorithm to find coeffient and intercept of an line.
    And also finds a cost function ( Mean Squarred Error )

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import math
```

```python
dataset = pd.read_csv('Salary_Data.csv')
dataset.head()
```

```python


def predict_using_sklean():
    df = pd.read_csv("Salary_Data.csv")
    r = LinearRegression()
    r.fit(df[['YearsExperience']],df.Salary)
    return r.coef_, r.intercept_

def gradient_descent(x,y):
    m_curr = 0
    b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.02

    cost_previous = 0

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n)*sum([value**2 for value in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if math.isclose(cost, cost_previous, rel_tol=1e-20):       # ( ie - 2 ) = 0.01
            break
        cost_previous = cost
        print ("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost, i))

    return m_curr, b_curr

if __name__ == "__main__":
    df = pd.read_csv("Salary_Data.csv")
    x = np.array(df.YearsExperience)
    y = np.array(df.Salary)

    m, b = gradient_descent(x,y)
    print(f"\nUsing gradient descent function: Coef {m}  Intercept {b}\n")

    m_sklearn, b_sklearn = predict_using_sklean()
    print(f"Using sklearn: Coef {m_sklearn}  Intercept {b_sklearn}")
```
